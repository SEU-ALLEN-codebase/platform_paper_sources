```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
%matplotlib inline

def ssp_projection_plot(ssp_proj_file):
    ssp_proj_stype=pd.read_csv(ssp_proj_file,sep=" ",header=None)
    ssp_proj_stype=np.array(ssp_proj_stype)
    ### delete column-L6b --> L6
    ssp_proj_stype2=np.delete(ssp_proj_stype,[5,12],axis=1)
    ssp_proj_stype2[:,4]=ssp_proj_stype2[:,4]+ssp_proj_stype[:,5]
    ssp_proj_stype2[:,10]=ssp_proj_stype2[:,10]+ssp_proj_stype[:,12]
    return ssp_proj_stype2
```

# SSp: projection plot over subtypes

```python
ssp_proj_stype_file = "/Users/krystal/Desktop/project/cloud_paper/1214_data/ssp_stype_proj.txt"
ssp_S=ssp_projection_plot(ssp_proj_stype_file)
```

```python
ssp_class=['SSp-bfd','SSp-ll','SSp-m','SSp-n','SSp-ul','SSp-un']
yticklabel1=['SSp-bfd(34)','SSp-ll(14)','SSp-m(57)','SSp-n(23)','SSp-ul(26)','SSp-un(4)']
xticklabel1=["L1","L2/3","L4","L5","L6","Non-\ncortical","L1","L2/3","L4","L5","L6a","Non-\ncortical"]

num_layer = 6
num_subtype = 6

X1=[0.5+i for i in range(num_layer)]
X2=[0.5+i for i in range(num_layer,num_layer*2)]
plt.figure(figsize=(12,6))
for i in range(num_layer):
    plt.scatter(X1,[(num_subtype-0.5-i) for j in range(num_layer)],s=ssp_S[i,:num_layer]*1000,c='cornflowerblue')    
    plt.scatter(X2,[(num_subtype-0.5-i) for j in range(num_layer)],s=ssp_S[i,num_layer:]*1000,c='lightcoral') 
    
plt.xlim(0,num_layer*2)
plt.ylim(0,num_subtype)
x = MultipleLocator(1)    # x10
y = MultipleLocator(1)    # y15
ax = plt.gca()
ax.xaxis.set_major_locator(x)
ax.yaxis.set_major_locator(y)
X=[i for i in range(num_layer*2)]
Y=[i+1 for i in range(num_subtype)]
plt.xticks(X,xticklabel1,fontsize=24)
plt.yticks(Y,yticklabel1,fontsize=24)

plt.grid(True, color='gray',which='major')
plt.savefig(f'/Users/krystal/Desktop/project/cloud_paper/1214_data/ssp_proj_R.png',bbox_inches='tight',pad_inches=0.2,dpi=600)
```
