```python
df_size = pd.read_csv('../assets/TeraDownsampleSize.csv')
brains_size = np.unique([i[0].split('\t')[0] for i in df_size.values]).tolist()
tmp = []
for i in brains_size:
    if i[0] in ['1','2']:
        tmp.append(int(i))
    else:
        tmp.append(i)
brains_size = tmp
print(len(brains_size),brains_size)
#ZhixiYun/data/registration/Inverse中205与之一致
```

```python
signal_dir = '/home/lyf/Research/cloud_paper/brain_statistics/statis_out/statis_out_corrected0315/'
modalities = os.listdir(signal_dir)
for modality in modalities:
    modality_dir = signal_dir+'/'+modality
    brains_signal = [b[:-4] for b in os.listdir(modality_dir) if b.endswith('.csv')]    
    brains_signal.sort()
    try:
        if brains_signal[0][0]=='1': 
            brains_signal = [int(b) for b in brains_signal]
    except:
        pass
    print(modality,len(brains_signal),brains_signal,)
    
    brains_inmeta = [b for b in brains_signal if b in brains_meta]
    brains_inlabel = [b for b in brains_signal if b in brains_zenglabel]
    brains_inyuanyuan = [b for b in brains_signal if b in brains_yuanyuan]
    brains_withsoma = [b for b in brains_signal if b in brains_soma]
    brains_intera = [b for b in brains_signal if b in brains_zengtera]
    brains_insize = [b for b in brains_signal if b in brains_size]
    
    print(f'in meta: {len(brains_inmeta)}    with label: {len(brains_inlabel)}    in yuanyuan: {len(brains_inyuanyuan)}    with soma: {len(brains_withsoma)}    in zengteracount: {len(brains_intera)}    in teracount: {len(brains_insize)}')
    if modality=='fMOST-Zeng':
        print(set(brains_zengtera)-set(brains_signal),'\n')
    else:
        print('\n')
```
