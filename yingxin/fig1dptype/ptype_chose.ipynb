```python
import pandas as pd
ptype_csvfile = '1741_Celltype_ptype.csv'
ptype_csv = pd.read_csv(ptype_csvfile,usecols=list(range(9)))
ptype_csv['Cortical_layer'][ptype_csv['Cortical_layer']=='2月3日']='2'
ptype_csv['Cortical_layer'] = ptype_csv['Cortical_layer'].fillna('nan')
print(ptype_csv.shape)
import numpy as np
ptype_csv_ = ptype_csv[ptype_csv['Brain_id']=='Ai82']
ptype_csv_ = ptype_csv_.drop(labels='Brain_id',axis=1)
print(ptype_csv_.shape)
ptype_csv = ptype_csv[ptype_csv['Brain_id']!='Ai82']
ptype_csv = ptype_csv.drop(labels='Celltype__',axis=1)
print(ptype_csv.shape)
ptype_csv_.columns = ptype_csv.columns
ptype_csv = pd.concat([ptype_csv_,ptype_csv],axis=0)
ptype_csv = ptype_csv.sort_index()
ptype_csv = ptype_csv.fillna(method='bfill',axis=1)
ptype_csv['Celltype'] = ptype_csv['Celltype'].fillna('nan')
ptype_csv['Celltype_'] = ptype_csv['Celltype_'].fillna('nan')
ptype_list = list(set(list((ptype_csv['Celltype']))))
print(len(ptype_list),ptype_list)
subptype_list = list(set(list((ptype_csv['Celltype_']))))
print(len(subptype_list),subptype_list)
layer_list = list(set(list((ptype_csv['Cortical_layer']))))
print(layer_list)
ptype_csv
```

```python
import os
ptype_dict = {}
for ptype in ptype_list:
    ptype_dict[ptype] = []
    for layer in layer_list:
        os.makedirs(ptype + '/' + layer, exist_ok=True)
        
regi25_swcdir = 'all/V2023_01_10/registration/S1_registered_from_zhixi'
s = 0
for i,row in ptype_csv.iterrows():
    name = row['Name']
    brain = row['Subclass_or_type']
    layer = row['Cortical_layer']
    ptype = row['Celltype']
    ptype_dict[ptype].append(name)
    subptype = row['Celltype_']
    
    regi25_swcfile = regi25_swcdir + '/' + brain + '/' + name + '_stps.swc'
    if os.path.exists(regi25_swcfile):
        regi25_ptype_swcfile = ptype + '/' + layer + '/' + subptype + '_' + name + '.swc'
        print(regi25_ptype_swcfile)
        s += 1
    else:
        print(f'\nNO SUCH {regi25_swcfile}\n')
    cp_str = f'cp {regi25_swcfile} {regi25_ptype_swcfile}'
    os.system(cp_str)    
print('\nplot num: ',s)
```
