```python
def write_swcobj_fromtrees(obj_des,trees):
    obj = []
    for tree in trees:
        df = pd.DataFrame(tree,columns=['n','t','x','y','z','r','p'])

        f = []
        k = 0
        id2line = {}
        for i in df.index.tolist():
            idx =  df.loc[i,'n']
            k = k+1
            id2line[idx] = str(k)
            s = ['v' ,str(df.loc[i,'x']),str(df.loc[i,'y']),str(df.loc[i,'z'])]
            f.append(s)
        for i in df.index.tolist():
            if df.loc[i, 'p'] not in id2line.keys():
                s = ['p', id2line[df.loc[i,'n']]]
                f.append(s)
                continue
            s = ['l', id2line[df.loc[i,'n']], id2line[df.loc[i, 'p']]]
            f.append(s)

        obj.append(f)

    length = 0
    new_v = []
    new_p = []
    new_l = []
    for f in obj:
        new_length = 0
        for leaf in f:
            if leaf[0]=='v': 
                new_v.append(leaf)
                new_length += 1
            elif leaf[0]=='p': 
                new_p.append(['p',float(leaf[1])+length])
            elif leaf[0]=='l': 
                new_l.append(['l',float(leaf[1])+length,float(leaf[2])+length])
        length += new_length

    with open(obj_des, 'w+') as fp:
        for i,leaf in enumerate(new_v):
            s = f'{leaf[0]} {leaf[1]} {leaf[2]} {leaf[3]}\n'
            fp.writelines(s)
        for i,leaf in enumerate(new_p):
            s = f'{leaf[0]} {leaf[1]}\n'
            fp.writelines(s)  
        for i,leaf in enumerate(new_l):
            s = f'{leaf[0]} {leaf[1]} {leaf[2]}\n'
            fp.writelines(s) 
            
    return 0
```

```python
def calc_score(module,module_feature_file='../../assets/corr.csv',agg=['mean','max','min']):
    feature_csv = pd.read_csv(module_feature_file)
    feature_csv.index = feature_csv.columns
    feature = (feature_csv.loc[module]).values.astype(float)
    feature_df = pd.DataFrame(feature.T,columns=module,index=feature_csv.columns)
    corr = feature_df.corr(min_periods=20, method='spearman')
    corr = corr.replace(1,np.nan)
    corr = abs(corr)
    data = []
    for e in agg:
        eva = eval(f'np.nan{e}(corr)')
        data.append([eva,e])    
    return data
    
plot = 0
for key in cross_intra_dict["intra"]:
    if len(module_example_dict[key])>=4:
        for r,regions in enumerate(module_regions_dict[key]):
            inters = set(regions) if r==0 else inters|set(regions)         
        module = module_dict[key]
        covers = inters&set(module)
        #if len(covers)==len(module):
        #if len(covers)>=len(module)-1:
        if 1:
            plot += 1    
            print('\n no.',plot,', module',key)
            struc = "same" if key in cross_intra_dict["intra"] else "different"
            print(module,struc,'structure','example number',len(module_example_dict[key]))
            module_csv[str(key)][14]=struc
            data = calc_score(module)
            module_csv[str(key)][11]=data[0][0]
            module_csv[str(key)][12]=data[1][0]
            module_csv[str(key)][13]=data[2][0]
            print('region diversity score',data)
```
